A decorator must accept a callable as its input argument
a decorator must return another callable

```python
def log(func):
    
    def wrapper(*args):
        print("Start demo...")
        func(*args)
        print("End of demo...")
    return wrapper
@log
def add(num1,num2,num3):
    print(num1+num2+num3)

@log
def div(num1,num2):
    print(num1/num2)
    
add(10,200,300)
div(30,70)
```
