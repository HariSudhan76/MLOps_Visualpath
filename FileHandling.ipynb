Run each line 1 by 1 without loading to a file

```python

def process(i,txt):
  return f"line: {i+1}- {txt}"

with open("/content/Filehandling.txt","r") as file:
  i=0
  while line := file.readline():
    print(process(i,line))
    i+=1
```

Handling Json file

```python
import json
def addition(a,b):
  return a+b

with open("/content/addition_num.txt","r") as num:
  for line in num:
    if line.strip():
      data = json.loads(line)
      print(f"Sum of {data['a']} & {data['b']} = {addition(data['a'],data['b'])}")


```
